Loop Through a List

Use the range() and len() functions to create a suitable iterable.



```python
thislist = ["apple","banana","cherry"]
for i in range(len(thislist)):
  print(thislist[i])
```

Using a While Loop
You can loop through the list items by using a while loop.

Use the len() function to determine the length of the list, then start at 0 and loop your way through the list items by referring to their indexes.

Remember to increase the index by 1 after each iteration.

```python
thislist = ["apple","banana","cherry"]
i = 0
while i<len(thislist):
  print(thislist[i])
  i = i+ 1
```

```python
x = [2, 3, 5, 6]

max_val = x[0]
next_max = x[0]

for i in x:
    if i > max_val:
        next_max = max_val
        max_val = i
    elif i > next_max and i != max_val:
        next_max = i

print("Sabse bada (Max):", max_val)
print("Doosra sabse bada (Next Max):", next_max)
```

```python
# Guessing game
import random
jackpot = random.randint(1,100)
chance = 1
print("<----You have a 5 chance --->")
print("<---- you guess number 1 to 100 ---->")
while chance<5:
  guess = int(input("Enter guess number: "))
  if guess >jackpot:
    print("Aur lower")
  elif guess <jackpot:
    print("upper number")
  else:
    print("you win")
    break
  chance += 1
else:
  print("you loss")
  print("Right guess number:",jackpot)
```
